```python
data = "/data_nfs/datasets/melc/melanoma/processed/training"
markers = list()
for sample in os.listdir(data):
    sample_path = os.path.join(data, sample)
    sample_markers = [ch.split("_")[1] for ch in os.listdir(sample_path) if ch.endswith(".tif")]
    for m in sample_markers:
        if "PBS" in m or "phase" in m or "ropidium" in m:
            continue
        if m not in markers:  
            if m.endswith("-FITC"):
                m = m[:-5]
            elif m.endswith("-PE"):
                m = m[:-3]
            markers.append(m)
markers = np.unique(markers)
```
